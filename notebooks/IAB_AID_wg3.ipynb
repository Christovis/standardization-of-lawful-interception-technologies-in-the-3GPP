```python
import numpy as np
import pandas as pd

import networkx as nx
from networkx.algorithms import bipartite

import pylab
from colour import Color
from pylab import cm
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use("publication_jcap.mplstyle")

from bigbang import listserv
from bigbang.analysis.listserv import ListservArchive
from bigbang.analysis.listserv import ListservList
from bigbang.visualisation import stackedareachart
from bigbang.visualisation import lines
from bigbang.visualisation import utils
from bigbang.analysis.utils import (
    get_index_of_msgs_with_subject,
    get_index_of_msgs_with_datetime,
)
from bigbang.visualisation import graphs
```

# 0 Load and Prepare the Data

```python
# load the dataset to associate stakeholders to sectors and nationality

df_organization_categories = pd.read_csv(
    "../2_IAB_AID/organization_categories.csv",
    sep=",",
    header=0,
    index_col=False,
)

# choose columns of interest
df_nation = df_organization_categories[[
    "name",
    "nationality",
    "email domain names",
]]
df_category = df_organization_categories[[
    "name",
    "category",  #"category",
    "email domain names",
]]

# filter out all non-empty cells
df_nation = df_nation.dropna()
df_category = df_category.dropna()

# create dictionary with domains as keys and nationality as value
df_nation_domains = {"email domain names": [], "nationality": []}
for i, row in df_nation.iterrows():
    domains = row["email domain names"].split(',')
    for dom in domains:
        df_nation_domains["email domain names"].append(dom.strip().lower())
        df_nation_domains["nationality"].append(row["nationality"])
df_nation_domains = pd.DataFrame.from_dict(df_nation_domains)

# create dictionary with domains as keys and stakeholdergroup as value
df_cat_domains = {"domain": [], "category": []}
for i, row in df_category.iterrows():
    domains = row["email domain names"].split(',')
    for dom in domains:
        df_cat_domains["domain"].append(dom.strip().lower())
        df_cat_domains["category"].append(row["category"])
df_cat_domains = pd.DataFrame.from_dict(df_cat_domains)
```

```python
# load mailing list
mlist_name = "3GPP_TSG_SA_WG6"
mlist = ListservList.from_mbox(
    name=mlist_name,
    filepath=f"../bigbang/archives/3GPP/{mlist_name}.mbox",
    include_body=True,
)
mlist.df = mlist.df.drop(
    columns=['Content-Transfer-Encoding', 'MIME-Version', 'message-id', 'Content-Type', 'organization']
)

# filter out domains that are also contained in df_nation_domains
mlist_nation = mlist.crop_by_address(
    header_field="from",
    per_address_field={"domain": np.unique(df_nation_domains["email domain names"].values)},
)

# filter out domains that are also contained in df_cat_domains
mlist_category = mlist.crop_by_address(
    header_field="from",
    per_address_field={"domain": np.unique(df_cat_domains["domain"].values)},
)
```

# 1 Communications Graphs

```python
# create di-graph with nationality as node attribues

nw = mlist_nation.get_sender_receiver_dict(
    address_field="domain",
)

mlist_nation.create_sender_receiver_digraph(
    nw=nw,
    node_attributes=df_nation_domains.rename(columns={"email domain names": "node_name", "nationality": "node_attribute"}),
)
```

```python
# create di-graph with category as node attribues

nw = mlist_category.get_sender_receiver_dict(
    address_field="domain",
)

mlist_category.create_sender_receiver_digraph(
    nw=nw,
    node_attributes=df_cat_domains.rename(columns={"domain": "node_name", "category": "node_attribute"}),
)
```

```python
selected_nodes = [
    name for name, attribute in mlist_nation.dg.nodes(data=True)
    if attribute['node_attribute'] != "-"
]
H = mlist_nation.dg.subgraph(selected_nodes)

selected_nodes = []
for node in H.nodes():
    in_selection = [
        1
        for in_edge in H.in_edges(node)
        if H[in_edge[0]][in_edge[1]]["weight"] > 35
    ]
    out_selection = [
        1
        for out_edge in H.out_edges(node)
        if H[out_edge[0]][out_edge[1]]["weight"] > 35
    ]
    if np.sum(in_selection + out_selection) > 0:
        selected_nodes.append(node)

F = mlist_nation.dg.subgraph(selected_nodes)


fig, axis = plt.subplots(
    1, 1,
    sharex=False, sharey=False,
    figsize=(8, 8),
    facecolor="w", edgecolor="w",
)

pos = nx.shell_layout(F)
node_size = graphs.node_size(F)
edges, edge_width = graphs.edge_thickness(F, entity_in_focus=[node for node in F.nodes()], exponent=1/3)

nx.draw_networkx_nodes(
    F,
    pos,
    node_color='b',
    node_size=node_size,
    #alpha=1/(node_size)**(1/4),
)

nx.draw_networkx_edges(
    F,
    pos,
    width=edge_width,
    edgelist=edges,
    edge_color=edge_width,
    edge_cmap=plt.cm.rainbow,
)

nx.draw_networkx_labels(
    F,
    pos,
    bbox={"facecolor":"w", "edgecolor":"w"},
)

plt.title(mlist_name)
plt.axis('off')

"""
plt.savefig(
    f"graphs_{mlist_name}_domain_names.png",
    format='png',
    transparent=True,
    dpi=300,
    bbox_inches='tight',
)#"""
```

# 1.1 Graph by Nationalities

```python
selected_nodes = [
    name for name, attribute in mlist_nation.dg.nodes(data=True)
    if attribute['node_attribute'] != "-"
]
H = mlist_nation.dg.subgraph(selected_nodes)

node_if = "huawei.com"

for nationality in df_nation_domains["nationality"].unique():
    selected_nodes = [
        name for name, attribute in H.nodes(data=True)
        if (attribute['node_attribute'] == nationality) and (name != node_if)
    ]
    if len(selected_nodes) > 1:
        for selected_node in selected_nodes[1:]:
            H = nx.contracted_nodes(H, selected_nodes[0], selected_node)

mapping = {}
for node in H.nodes(data=True):
    if node != node_if:
        mapping[node[0]] = node[1]["node_attribute"]
mapping[node_if] = node_if
        
H = nx.relabel_nodes(H, mapping)
```

```python
fig, axis = plt.subplots(
    1, 1,
    sharex=False, sharey=False,
    figsize=(8, 8),
    facecolor="w", edgecolor="w",
)

pos = nx.shell_layout(H)
node_size = graphs.node_size(H)
edges, edge_width = graphs.edge_thickness(
    H,
    entity_in_focus=[node for node in H.nodes()],
    exponent=1/2,
)

nx.draw_networkx_nodes(
    H,
    pos,
    node_color='b',
    node_size=node_size,
    #alpha=1/(node_size)**(1/4),
)

nx.draw_networkx_edges(
    H,
    pos,
    width=edge_width,
    edgelist=edges,
    edge_color=edge_width,
    edge_cmap=plt.cm.rainbow,
)

nx.draw_networkx_labels(
    H,
    pos,
    bbox={"facecolor":"w", "edgecolor":"w"},
)
nx.draw_networkx_labels(
    H.subgraph(node_if),
    {node_if: pos[node_if]},
    bbox={"facecolor":"w", "edgecolor":"k"},
)

plt.title(mlist_name)
plt.axis('off')

"""
plt.savefig(
    f"graphs_{mlist_name}_domain_nationality.png",
    format='png',
    transparent=True,
    dpi=300,
    bbox_inches='tight',
)#"""
```

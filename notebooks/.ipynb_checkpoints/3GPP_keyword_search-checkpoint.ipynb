```python
import datetime
import pandas as pd
import numpy as np

from colour import Color
from pylab import cm
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

from bigbang.analysis.listserv import ListservMailList
from bigbang.bigbang_io import pandas_dataframe_to_mbox

plt.style.use("./publication_jcap.mplstyle")
```

```python
# Selection
df_qus = pd.read_csv(
    #"../data/bigrams_unsupervised_verified_in_maillist.csv",
    "../data/bigrams_unsupervised_verified_in_maillist_attachments.csv",
    header=0,
    index_col=0,
)
df_qus.loc[:, "message-date"].update(
    df_qus.loc[:, "message-date"].apply(
        lambda x: datetime.datetime.strptime(x, "%a, %d %b %Y %H:%M:%S %z")
    )
)
non_query_columns = [
    #'message-body',
    'message-attachment',
    'message-id',
    'message-date',
    'message-from',
    'archived-at',
]
dtm = df_qus.loc[:, ~df_qus.columns.isin(non_query_columns)]

# Mailing list
mlist_name = "3GPP_TSG_SA_WG3_LI"
filepath = f"/home/christovis/InternetGov/bigbang/archives/3GPP/{mlist_name}.mbox"
mlist = ListservMailList.from_mbox(
    name=mlist_name,
    filepath=filepath,
)

print(len(df_qus), len(dtm), len(mlist.df))
```

```python
counts = (dtm != 0).sum(axis=0)
rank = np.argsort(counts)
query_count = {'query': [], 'count': []}
for col, cou in zip(dtm.columns[rank], counts[rank]):
    query_count['query'].append(col)
    query_count['count'].append(cou)
query_count = pd.DataFrame.from_dict(query_count)
```

```python
indices = []
for column in dtm.columns:
    df_qu = df_qus[df_qus[column] != 0]
    min_datetime = df_qu['message-date'].min()
    index = df_qus[df_qus['message-date'] == min_datetime].index[0]
    indices.append(index)
    
df_qus_first = df_qus.loc[indices]
mlist_first = ListservMailList.from_pandas_dataframe(
    df=mlist.df[mlist.df['message-id'].isin(df_qus_first['message-id'].values)],
    name=mlist_name,
    filepath=filepath,
)
```
